```python
#Преобразование датасета в датасет состоящий из пар: для одинакого query_id первый элемент более выжен чем второй.
def rank_queryId_feature(dataset):
    ranks = dataset['rank'].to_numpy()
    queryIds = dataset['query_id'].to_numpy()
    features = dataset.iloc[:, 2:].to_numpy()
    return ranks, queryIds, features

def get_pair_doc_data(ranks, queryIds):
     pairs = []
     tmp_x0 = []
     tmp_x1 = []
     for i in tqdm(range(0, len(queryIds) - 1), total=len(queryIds)):
         for j in range(i + 1, len(queryIds)):
             if queryIds[i] != queryIds[j]:
                 break
             #Use document pairs with different relevance
             if (queryIds[i] == queryIds[j]) and (ranks[i] != ranks[j]):
                 #Put the most relevant first, and keep the first doc in the document pair more relevant to the query than the second doc
                 if ranks[i] > ranks[j]:
                     pairs.append([i,j])
                 else:
                     pairs.append([j,i])
     print('fond {} doc pairs'.format(len(pairs)))
     return len(pairs), pairs
```

```python
class Dataset(data.Dataset):
     def __init__(self, df):
         ranks, queryIds, self.features = rank_queryId_feature(df)
         #  pair combination
         self.datasize, self.pairs = get_pair_doc_data(ranks, queryIds)
         print(self.datasize)

     def __getitem__(self, index):
         data1 = torch.from_numpy(self.features[self.pairs[index][0]]).float()
         data2 = torch.from_numpy(self.features[self.pairs[index][1]]).float()
         return data1, data2

     def __len__(self):
         return self.datasize
```

```python
class test_val_Dataset(data.Dataset):
     def __init__(self, df):
         self.df = df
         dictionary = df.groupby('query_id').groups
         self.indexes = [np.array(value) for key, value in dictionary.items() if len(value) > 1]


     def __getitem__(self, index):
         return torch.from_numpy(np.array(self.df.loc[self.indexes[index]].iloc[:, 2:])), torch.from_numpy(np.array(self.df.loc[self.indexes[index]]['rank']))

     def __len__(self):
         return len(self.indexes)
```

```python
def get_loader(df, batch_size, shuffle):
    data_loader = torch.utils.data.DataLoader(
        dataset=df,
        batch_size = batch_size,
        shuffle = shuffle,
    )
    return data_loader
```



# Обучение



```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import os

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

class RankNet(nn.Module):
    def __init__(self, inputs, hidden_size, outputs):
        super(RankNet, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(inputs, hidden_size),
            nn.Linear(hidden_size, hidden_size),
            nn.ReLU(inplace=True),
            nn.Linear(hidden_size, outputs),
        )
        self.sigmoid = nn.Sigmoid()

    def forward(self, input_1, input_2):
        result_1 = self.model(input_1) #Predict input_1 score
        result_2 = self.model(input_2) #Predict input_2 score
        pred = self.sigmoid(result_1 - result_2) #input_1 is more relevant than input_2
        return pred

    def predict(self, input):
        result = self.model(input)
        return result
```

```python
from sklearn.metrics import ndcg_score


def compute_metrics(
    score: torch.Tensor,
    true: torch.LongTensor,
):
    """
    Args:
        outputs: the model outputs (batch_size, num_classes, sequence_len)
        labels: the correct classes (batch_size, sequence_len)

    Returns:
        metrics: mapping metric names to their corresponding values
    """

    metrics = {}

    true = true.to('cpu').detach().numpy()
    score = score.to('cpu').detach().squeeze(-1).numpy()
    metrics['ndcg'] = ndcg_score(
        y_true=true,
        y_score=score,
    )

    metrics['ndcg_5'] = ndcg_score(
        y_true=true,
        y_score=score,
        k=5
    )

    return metrics
```

```python
def train_epoch(
    model: torch.nn.Module,
    dataloader: torch.utils.data.DataLoader,
    optimizer: torch.optim.Optimizer,
    criterion: torch.nn.Module,
    device: torch.device,
    epoch: int,
) -> None:


    model.train()

    epoch_loss = []


    for i, (data1, data2) in tqdm(
        enumerate(dataloader),
        total=len(dataloader),
        desc="loop over train batches",
    ):
        data1, data2 = data1.to(device), data2.to(device)

        outputs = None
        loss = None

        optimizer.zero_grad()
        pred = model(data1, data2)

        label_size = data1.size()[0]
        loss = criterion(pred, torch.from_numpy(np.ones(shape=(label_size, 1))).float().to(device))
        loss.backward()
        optimizer.step()


        epoch_loss.append(loss.item())



    avg_loss = np.mean(epoch_loss)
    print(f"Train loss: {avg_loss}\n")
```

```python
def evaluate_epoch(
    model: torch.nn.Module,
    dataloader: torch.utils.data.DataLoader,
    criterion: torch.nn.Module,
    device: torch.device,
    epoch: int,
) -> None:

    model.eval()

    epoch_loss = []
    batch_metrics_list = {
        'ndcg' : [],
        'ndcg_5': []
    }

    with torch.no_grad():

        for i, (input, ranks) in tqdm(
            enumerate(dataloader),
            total=len(dataloader),
            desc="loop over test batches",
        ):
            input, ranks = input.to(device), ranks.to(device)
            outputs = model.predict(input.float())


            batch_metrics = compute_metrics(
                score=outputs,
                true=ranks,
            )

            for metric_name, metric_value in batch_metrics.items():
                batch_metrics_list[metric_name].append(metric_value)

        for metric_name, metric_value_list in batch_metrics_list.items():
            metric_value = np.mean(metric_value_list)
            print(f"Test {metric_name}: {metric_value}\n")

        return np.mean(batch_metrics_list['ndcg_5'])
```

```python
from torch.optim.lr_scheduler import StepLR

def train(
    n_epochs: int,
    model: torch.nn.Module,
    train_dataloader: torch.utils.data.DataLoader,
    valid_dataloader: torch.utils.data.DataLoader,
    optimizer: torch.optim.Optimizer,
    criterion: torch.nn.Module,
    device: torch.device,
) -> None:

    scheduler = StepLR(optimizer, step_size=3, gamma=0.5)
    best_score = 0
    for epoch in range(n_epochs):

        print(f"Epoch [{epoch+1} / {n_epochs}]\n")

        train_epoch(
            model=model,
            dataloader=train_dataloader,
            optimizer=optimizer,
            criterion=criterion,
            device=device,
            epoch=epoch,
        )

        cur_score = evaluate_epoch(
            model=model,
            dataloader=valid_dataloader,
            criterion=criterion,
            device=device,
            epoch=epoch,
        )
        scheduler.step()
        cur_lr = scheduler.get_last_lr()
        print(f'current_lr = {cur_lr}')
        if cur_score > best_score:
             best_score = cur_score
             torch.save(model.state_dict(), f"best_model_{epoch}.pth")
```

```python
model = RankNet(inputs=144, hidden_size=10, outputs=1).to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=0.2)
criterion = torch.nn.CrossEntropyLoss()
```

```python
train(
    n_epochs=2,
    model=model,
    train_dataloader=train_dataloader,
    valid_dataloader=val_dataloader,
    optimizer=optimizer,
    criterion=criterion,
    device=device,
)
```
